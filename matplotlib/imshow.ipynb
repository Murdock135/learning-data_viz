Example showing default norm vs norm = asinh

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample 2D array (image or heatmap)
data = np.random.rand(10, 10) * 100  # Random data scaled to 0-100

# Display the original data
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Original Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display data with vmin and vmax specified
plt.subplot(1, 3, 2)
plt.imshow(data, cmap='viridis', aspect='auto', vmin=20, vmax=80)
plt.colorbar()
plt.title('With vmin=20 and vmax=80')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display data with linear normalization using norm='linear'
plt.subplot(1, 3, 3)
plt.imshow(data, cmap='viridis', aspect='auto', norm='asinh', vmin=20, vmax=80)
plt.colorbar()
plt.title("With norm='linear', vmin=20, vmax=80")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

Example showing default vmin, vmax and specified values for vmin, vmax

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample 2D array (image or heatmap)
data = np.random.rand(10, 10) * 100  # Random data scaled to 0-100

# Display the original data with default vmin and vmax
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(data, cmap='viridis', aspect=3.2)
plt.colorbar()
plt.title('Default vmin and vmax, aspect=3.2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display data with explicit vmin and vmax
plt.subplot(1, 2, 2)
plt.imshow(data, cmap='viridis', aspect='auto', vmin=20, vmax=80)
plt.colorbar()
plt.title('With vmin=20 and vmax=80')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
